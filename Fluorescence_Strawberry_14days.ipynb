```python
import pandas as pd
import numpy as np
np.set_printoptions(suppress=True)
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, peak_prominences
from scipy.signal import cspline1d, cspline1d_eval
%matplotlib inline
```

```python
data = pd.read_csv("./strawberryLC.csv")
data.head(6)
```

```python
data1 = data['point_1'].values
data2 = data['point_2'].values
data3 = data['point3'].values
```

```python
peaks1, _ = find_peaks(data1, distance=400)
peaks2, _ = find_peaks(data2, height=20)
peaks3, _ = find_peaks(data3, height=15)
```

```python
plt.plot(data1, label='2nd Day')
plt.plot(data2, label='4th Day')
plt.plot(data3, label='6th Day')
plt.plot(peaks1, data1[peaks1], 'x', label='peaks')
plt.rcParams["figure.figsize"] = (14,10)
plt.legend(loc="lower center", prop={'size': 12})
plt.title('Rapid Ligth Curve 90 second duration - 30 min Dark Adaptation - 300 μmoles-s-m2 - 14 hours Photoperiod')
plt.xlabel('Frames')
plt.ylabel('Fluorescence (A.U.)')
```
